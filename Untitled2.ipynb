```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
import uuid
import tensorflow as tf
import keras

from glob import glob
import cv2

# import model 
from keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
from keras.models import Sequential 
from keras.layers import Conv2D, MaxPooling2D, Activation, Dropout, Flatten, Dense, GaussianNoise, Rescaling

#import warnings
import warnings
warnings.filterwarnings('ignore')

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
# for dirname, _, filenames in os.walk('datasets/fruits-360'):
#     for filename in filenames:
#         print(os.path.join(dirname, filename))
        
# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session        
```

```python
train_dir = './datasets/fruits-360/Training/'
test_dir =  './datasets/fruits-360/Test/'
```

```python
fruits = []
fruits_image = []
for i in os.listdir(train_dir):
    for image_filename in os.listdir(train_dir + i):
        fruits.append(i) # name of the fruit 
        fruits_image.append(i + '/' + image_filename) # image of the fruit
```

```python
train_fruits = pd.DataFrame(fruits, columns=["Fruits"])
train_fruits["Fruits Image"] = fruits_image
```

```python
train_fruits
```

```python
from collections import Counter
newData = Counter(train_fruits["Fruits"])

all_fruits = newData.most_common()

classes,_ = zip(*all_fruits)
classes,_ = list(classes), _
```

```python
most_fruits = newData.most_common(15)

print("The 15 most frequently found fruits in the data set and their number")
most_fruits
```

```python
x,y = zip(*most_fruits)
x,y = list(x),list(y)
```

```python
print("The 15 most frequently found fruits in the dataset -- > \n ", x)
```

```python
plt.figure(figsize=(15,10))
ax= sns.barplot(x=x, y=y, palette=sns.color_palette("BuGn_r", 15)) 
plt.xlabel('Fruits', size = 20) 
plt.ylabel('Number of Fruits', size = 20 )
plt.xticks(rotation = 75)
plt.title('The 15 most frequently found fruits in the data set and their number') 
plt.show()
```

```python
plt.figure(figsize=(25,16))
how_many_fruits = len(x)

for i in range(how_many_fruits):
    fruits = train_fruits[train_fruits["Fruits"] == x[i]]["Fruits Image"].values [1]
    
    plt.subplot(3,5,i+1)
    img = load_img(train_dir + fruits)
    plt.imshow(img)
    plt.title(x[i].upper(), color = "green", fontsize = 15 , fontweight = 600)
    plt.axis("off")

plt.suptitle("The 15 Most Abundant Fruits", fontsize = 25 , color = "darkred", fontweight = 'bold')
plt.show()
```

```python
array_image = img_to_array(img)
print("Image Shape --> ", array_image.shape)
```

```python
fruitCountUnique = glob(train_dir + '/*' )
numberOfClass = len(fruitCountUnique)
print("How many different fruits are there  -->  ",numberOfClass)
```

```python
model = Sequential()
model.add(Rescaling(1./255))

model.add(Conv2D(16, 3, padding='same', activation='relu'))
model.add(MaxPooling2D())

model.add(Conv2D(32, 3, padding='same', activation='relu'))
model.add(MaxPooling2D())

model.add(Conv2D(64, 3, padding='same', activation='relu'))
model.add(MaxPooling2D())

model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(numberOfClass))
model.add(Activation("softmax"))
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='rmsprop',
              metrics = ["accuracy"])
```

```python
epochs = 30
batch_size = 32
```

```python
print("Target Size --> ", array_image.shape[:2])
```

```python
import random

# Add random noise to an image
def add_noise(img):
    VARIABILITY = 5
    deviation = VARIABILITY*random.random()
    noise = np.random.normal(0, deviation, img.shape)
    img += noise
    np.clip(img, 0., 255.)
    return img
```

```python
train_datagen = ImageDataGenerator(
    rescale= 1./255,
    shear_range = 0.3,
    horizontal_flip=True,
    zoom_range = 0.3,
    width_shift_range=0.3,
    height_shift_range=0.3,
    rotation_range=45/2,
    brightness_range=[0.4,1.0],
#     preprocessing_function=add_noise
)

test_datagen = ImageDataGenerator(rescale= 1./255)
```

```python
train_generator = train_datagen.flow_from_directory(
        train_dir, 
        target_size= array_image.shape[:2],
        batch_size = batch_size,
        color_mode= "rgb",
        class_mode= "categorical"
)
```

```python
test_generator = test_datagen.flow_from_directory(
        test_dir, 
        target_size= array_image.shape[:2],
        batch_size = batch_size,
        color_mode= "rgb",
        class_mode= "categorical"
)
```

```python
fig = plt.figure(figsize=(10, 10))
columns = 4
rows = 4

for i in range(1, columns*rows +1):    
    img, label = train_generator.next()
    fig.add_subplot(rows, columns, i)
    plt.imshow((img[0] * 255).astype(np.uint8))
plt.show()
```

```python
for data_batch, labels_batch in train_generator:
    print("data_batch shape --> ",data_batch.shape)
    print("labels_batch shape --> ",labels_batch.shape)
    break
```

```python
hist = model.fit_generator(
        generator = train_generator,
        steps_per_epoch = 1600 // batch_size,
        epochs=epochs,
        validation_data = test_generator,
        validation_steps = 800 // batch_size)
```

```python
hist.history.keys()
```

```python
plt.figure(1)  
   
 # summarize history for accuracy  
   
plt.subplot(211)  
plt.plot(hist.history['accuracy'])  
plt.plot(hist.history['val_accuracy'])  
plt.title('model accuracy')  
plt.ylabel('accuracy')  
plt.xlabel('epoch')  
plt.legend(['train', 'test'], loc='upper left')  
   
 # summarize history for loss  
   
plt.subplot(212)  
plt.plot(hist.history['loss'])  
plt.plot(hist.history['val_loss'])  
plt.title('model loss')  
plt.ylabel('loss')  
plt.xlabel('epoch')  
plt.legend(['train', 'test'], loc='upper left')  
plt.show()
```
